```python
import pandas as pd
import json
import pprint
```

```python
imdb_df = pd.read_csv("resources/imdb_top_1000.csv")
imdb_df.head()
```

```python
title_list = list()

with open("resources/movie_ids.txt", 'r', encoding = "utf8") as file:
    for line in file:
        j = json.loads(line)
        title_list.append(j['original_title'])

pprint.pp(title_list)
```

```python
for i, name in imdb_df.Series_Title.items():
    if name not in title_list:
        print(name)
```
