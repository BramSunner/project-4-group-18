```python
# Dependencies.
# Data.
import pandas as pd
import numpy as np

# Helper functions.
from feature_processor import lemmatizeAssist, processText, columnMerger 
```

```python
# Load dataset.
df = pd.read_csv("resources/movie_dataset.csv")

# Select rows needed.
df = df[['title', 'rating', 'genre', 'director', 'cast1', 'cast2', 'cast3', 'cast4', 'overview']]

# Display.
display(df.info())
display(df.head())
```

```python
# Drop duplicate titles.
c_df = df.drop_duplicates('title')

c_df.info()
```

```python
# Change Sci-Fi to Science Fiction in genre.
c_df['genre'] = c_df.genre.apply(lambda x: x.replace("Sci-Fi", "Science Fiction"))

c_df.info()
```

```python
# Smush the overview.
c_df['lemma_text'] = c_df.overview.apply(lambda x: processText(x))

display(c_df.head())
```

```python
# Smush all columns together.
c_df['feature'] = c_df.apply(lambda row: columnMerger(row), axis = 1)

c_df.head()
```

```python
# Grab columns.
c_df = c_df[['title', 'rating', 'feature']]

c_df.head()
```

```python
# It's done.
c_df.to_csv("resources/ml_movie.csv", index = False)
```
